# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def urun_ekle():
    # Kullanıcıdan ürün bilgilerini al
    urun_ad = input("Ürün adını girin: ")
    urun_fiyat = input("Ürün fiyatını girin: ")

    # Dosyayı aç ve ürün bilgilerini dosyaya yaz
    with open("urunler.txt", "a") as file:
        file.write(urun_ad + "," + urun_fiyat + "\n")

    print("Ürün başarıyla eklendi!")
````

1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def kelime_degistir(filename, eski_kelime, yeni_kelime):
    with open(filename, 'r') as file:
        icerik = file.read()

    # Eski kelimeyi yeni kelimeyle değiştir
    icerik = icerik.replace(eski_kelime, yeni_kelime)

    with open(filename, 'w') as file:
        file.write(icerik)

    print(f"{eski_kelime} kelimesi {yeni_kelime} ile değiştirildi.")

kelime_degistir('ornek.txt', 'eski_kelime', 'yeni_kelime')

````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

    def __str__(self):
        return f"Name: {self.name} Speed: {self.speed} Mileage: {self.mileage}"


class Bus(Vehicle):
    def __init__(self, name, speed, mileage, category):
        super().__init__(name, speed, mileage)
        self.category = category

    def __str__(self):
        return super().__str__() + f" Category: {self.category}"


````
